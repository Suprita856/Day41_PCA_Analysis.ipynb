INTRODUCTION :-

Principal Component Analysis (PCA) is a dimensionality reduction technique used to simplify high-dimensional datasets while retaining as much important information as possible. In many real-world datasets, features can be redundant or noisy. PCA addresses this by transforming the original features into a smaller set of new, uncorrelated variables called principal components, which capture the maximum variance in the data.

OBJECTIVES :-

1.Reduce the number of features in a dataset
2.Retain maximum information (variance)
3.Eliminate redundancy and multicollinearity
4.Improve model efficiency and performance
5.Enable better visualization of high-dimensional data.

KEY POINTS :-

1.PCA transforms original features into linear combinations
2.Components are orthogonal (uncorrelated)
3.Variance = Information in PCA context
4.First component (PC1) captures the highest variance
5.Scree Plot helps choose optimal number of components
6.Standardization is essential before applying PCA

Advantages

1.Reduces dimensionality and computation cost
2.Helps prevent overfitting
3.Removes multicollinearity
4.Improves data visualization
5.Captures most important patterns in fewer dimensions

Disadvantages

1.Loss of interpretability (components are not human-readable)
2.Possible loss of information if too many dimensions are removed
3.Sensitive to scaling (requires standardization)
6.Assumes linear relationships in data

Conclusion

PCA is a powerful technique for simplifying complex datasets while preserving their essential structure. In this task, reducing 13 features to 3 components while retaining 96% variance demonstrated how effectively PCA can compress data without significant information loss. Although it introduces a trade-off in interpretability, it greatly enhances computational efficiency and visualization, making it an essential tool in data analysis and machine learning.
